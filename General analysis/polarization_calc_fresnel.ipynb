```python
import numpy as np
import matplotlib.pyplot as plt

import pypolar.fresnel as fresnel

DEFAULT_RENDERER = 'jupyterlab'
```

```python
# C60
m = complex(2.5, 1.25)
# m = 1.5
theta=45

Ts = fresnel.T_per(m, theta, deg=True)
Tp = fresnel.T_par(m, theta, deg=True)

print(f'The absorption of s-polarized light is {np.round(Ts/Tp, 3)*100}% that of p-polarized light in bulk C60.')
```

```python
Rs = fresnel.R_per(m, theta, deg=True)
Rp = fresnel.R_par(m, theta, deg=True)

print(f'(1-Rs)/(1-Rp) = {np.round((1-Rs)/(1-Rp), 3)*100}%')
```

If the beam didn't change cross-sectional area across the boundary between mediums, would these two calculations give the same result? i.e. Would Ts = 1-Rs? It seems like it (based on using m=1.5).
